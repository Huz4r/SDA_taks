sesion_01_20

```python
df_air = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Sensors/sesion_01_20/sensor01_air.txt', header=None, names=['X', 'Y'])
df_izopropanol = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Sensors/sesion_01_20/sensor01_izopropanol.txt', header=None, names=['X', 'Y'])
df_water = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Sensors/sesion_01_20/sensor01_water.txt', header=None, names=['X', 'Y'])

# Print basic information about each data frame

print("df_air info:")
print(df_air.info())
print("df_air describe:")
print(df_air.describe())
print("==============")
print("df_izopropanol info:")
print(df_izopropanol.info())
print("df_izopropanol describe:")
print(df_izopropanol.describe())
print("==============")
print("df_water info:")
print(df_water.info())
print("df_water describe:")
print(df_water.describe())
print("==============")

fig = plt.figure(figsize=(16, 10))
plt.plot(df_air['X'], df_air['Y'], label='air')
plt.plot(df_water['X'], df_water['Y'], label='water')
plt.plot(df_izopropanol['X'], df_izopropanol['Y'], label='izopropanol')

plt.xlabel(r'$\lambda$ [nm]')
plt.ylabel('P/Pref [dB]')
plt.grid(True)
plt.legend()

plt.show()
```

Zakładam, że w każdym pomiaże kolumna z danymi dotyczącymi długości fali światła ma takie same wartości. Poniższy skrypt ma potwierdzić czy to założenie jest prawdziwe.

```python
from google.colab import drive
import os
drive.mount('/content/drive')


# Define the folder path and file extension
folder_path = "/content/drive/MyDrive/Colab Notebooks/Sensors/"
extension = ".txt"

# Create a dictionary to store the second column values for each file
values_dict = {}

# Loop through all files in the folder
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith(extension):
            file_path = os.path.join(root, file)
            with open(file_path, "r") as f:
                lines = f.readlines()
                # Extract the second column values and store them in a list
                try:
                    values = [float(line.strip().split(",")[1]) for line in lines]
                    values_dict[file] = values
                except (ValueError, IndexError):
                    print(f"Skipping file {file_path} because it does not have valid data.")

# Check if the second column values are the same for all files
if len(values_dict) > 0:
    first_file_values = values_dict[list(values_dict.keys())[0]]
    same_values = True
    for values in values_dict.values():
        if values != first_file_values:
            same_values = False
            break
    if same_values:
        print("The second column values are the same for all files.")
    else:
        print("The second column values are not the same for all files.")
else:
    print("No files were found or none of the files contain valid data.")
```

```python
# Adding headings

import os

# Define input and output folders
input_folder = '/content/drive/MyDrive/Colab Notebooks/Sensors'
output_folder = '/content/drive/MyDrive/Colab Notebooks/Sensors_merge/Sessions_with_headers'

# Loop over each file in the input folder
for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith('.txt'):
            # Read in the data from the input file
            file_path = os.path.join(root, file)
            with open(file_path, 'r') as f:
                data = f.read().splitlines()
            
            # Add headers to the data
            header = f'{file.split(".")[0]}_amplitude,{file.split(".")[0]}_wavelength\n'
            data.insert(0, header)
            
            # Write the data to the output file
            output_file_path = os.path.join(output_folder, file)
            with open(output_file_path, 'w') as f:
                f.write('\n'.join(data))
            
            # Verify that the output file exists
            if os.path.exists(output_file_path):
                print(f'Successfully saved {output_file_path}')
            else:
                print(f'Error saving {output_file_path}')



```

```python
# import libraries
%matplotlib inline

from google.colab import drive
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

drive.mount('/content/drive')

# load the dataset with amplitude and wavelength measurements in air, isopropanol, and water
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Sensors_merge/one_file_to_rule_them_all.txt")

# extract input and output variables
X_air = df[['sensor01_air_amplitude', 'sensor01_air_wavelength']].values
y_isoprop = df[['sensor01_izopropanol_amplitude', 'sensor01_izopropanol_wavelength']].values
y_water = df[['sensor01_water_amplitude', 'sensor01_water_wavelength']].values

# define polynomial degree
degree = 6

# create polynomial features
poly = PolynomialFeatures(degree)
X_air_poly = poly.fit_transform(X_air)

# define k-fold parameters
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# initialize lists to store evaluation metrics
mse_list = []
r2_list = []
mae_list = []
rmse_list = []
mape_list = []

# loop through k-fold splits
for train_index, test_index in kf.split(X_air_poly):
  
  # split data into training and testing sets
  X_train, X_test = X_air_poly[train_index], X_air_poly[test_index]
  y_isoprop_train, y_isoprop_test = y_isoprop[train_index], y_isoprop[test_index]
  y_water_train, y_water_test = y_water[train_index], y_water[test_index]

  # fit polynomial regression model to training data
  lin_reg_isoprop = LinearRegression()
  lin_reg_isoprop.fit(X_train, y_isoprop_train)

  lin_reg_water = LinearRegression()
  lin_reg_water.fit(X_train, y_water_train)

  # make predictions on testing data
  y_isoprop_pred = lin_reg_isoprop.predict(X_test)
  y_water_pred = lin_reg_water.predict(X_test)

  # evaluate predictions using different metrics
  mse_isoprop = mean_squared_error(y_isoprop_test, y_isoprop_pred)
  r2_isoprop = r2_score(y_isoprop_test, y_isoprop_pred)
  mae_isoprop = mean_absolute_error(y_isoprop_test, y_isoprop_pred)
  rmse_isoprop = np.sqrt(mse_isoprop)
  mape_isoprop = np.mean(np.abs((y_isoprop_test - y_isoprop_pred) / y_isoprop_test)) * 100

  mse_water = mean_squared_error(y_water_test, y_water_pred)
  r2_water = r2_score(y_water_test, y_water_pred)
  mae_water = mean_absolute_error(y_water_test, y_water_pred)
  rmse_water = np.sqrt(mse_water)
  mape_water = np.mean(np.abs((y_water_test - y_water_pred) / y_water_test)) * 100

  # append metrics to lists
  mse_list.append([mse_isoprop, mse_water])
  r2_list.append([r2_isoprop, r2_water])
  mae_list.append([mae_isoprop, mae_water])
  rmse_list.append([rmse_isoprop, rmse_water])
  mape_list.append([mape_isoprop, mape_water])

# convert metrics lists to numpy arrays
mse_arr = np.array(mse_list)
r2_arr = np.array(r2_list)
mae_arr = np.array(mae_list)
rmse_arr = np.array(rmse_list)
mape_arr = np.array(mape_list)

# calculate mean and standard deviation of metrics
mse_mean = np.mean(mse_arr, axis=0)
mse_std = np.std(mse_arr, axis=0)

r2_mean = np.mean(r2_arr, axis=0)
r2_std = np.std(r2_arr, axis=0)

mae_mean = np.mean(mae_arr, axis=0)
mae_std = np.std(mae_arr, axis=0)

rmse_mean = np.mean(rmse_arr, axis=0)
rmse_std = np.std(rmse_arr, axis=0)

mape_mean = np.mean(mape_arr, axis=0)
mape_std = np.std(mape_arr, axis=0)

# print evaluation metrics
print("Isopropanol: ")
print("MSE: {:.2f} +/- {:.2f}".format(mse_mean[0], mse_std[0]))
print("R2 Score: {:.2f} +/- {:.2f}".format(r2_mean[0], r2_std[0]))
print("MAE: {:.2f} +/- {:.2f}".format(mae_mean[0], mae_std[0]))
print("RMSE: {:.2f} +/- {:.2f}".format(rmse_mean[0], rmse_std[0]))
print("MAPE: {:.2f}% +/- {:.2f}%".format(mape_mean[0], mape_std[0]))

print("\nWater: ")
print("MSE: {:.2f} +/- {:.2f}".format(mse_mean[1], mse_std[1]))
print("R2 Score: {:.2f} +/- {:.2f}".format(r2_mean[1], r2_std[1]))
print("MAE: {:.2f} +/- {:.2f}".format(mae_mean[1], mae_std[1]))
print("RMSE: {:.2f} +/- {:.2f}".format(rmse_mean[1], rmse_std[1]))
print("MAPE: {:.2f}% +/- {:.2f}%".format(mape_mean[1], mape_std[1]))

# plot actual vs predicted values
fig = plt.figure(figsize=(16, 10))

plt.plot(y_isoprop_pred[:,0], y_isoprop_pred[:,1], color='blue', label="Isopropanol (Predicted)")
plt.plot(y_isoprop_test[:,0], y_isoprop_test[:,1], color='red', label="Isopropanol (Real)")
plt.xlabel(r'$\lambda$ [nm]')
plt.ylabel('P/Pref [dB]')
plt.grid(True)
plt.legend()
plt.title('Comparison of Actual Isopropanol Substance Parameters to Predicted Values')
plt.show()

fig = plt.figure(figsize=(16, 10))

plt.plot(y_water_pred[:,0], y_water_pred[:,1], color='blue', label="Water (Predicted)")
plt.plot(y_water_test[:,0], y_water_test[:,1], color='red', label="Water (Real)")
plt.xlabel(r'$\lambda$ [nm]')
plt.ylabel('P/Pref [dB]')
plt.grid(True)
plt.legend()
plt.title('Comparison of Actual Water Substance Parameters to Predicted Values')
plt.show()

plt.tight_layout()
plt.show()
```
